```python
#importing library
import pandas as pd
```

```python
#create dataframe
df=pd.DataFrame({
    'Name':['Bhavya','Reema','Anjali','Janvi'],
    'Maths':[100,80,90,95],
    'Science':[90,94,95,99],
    'English':[95,98,92,100],
    'Computer Science':[100,70,99,100]
})
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# calculating total marks
df['Total Marks']=df['Maths'] + df['English'] + df['Science']+df['Computer Science']
```

```python
#show data frame
df
```

```python
# calculating percentage
df['Percentage']=df['Total Marks'] / 4
```

```python
# show dataframe
df
```

```python
# Assign Grades
def grades(percentage):
    if percentage>=95:
        return 'A+'
    elif percentage>=90:
        return 'A'
    elif percentage>=85:
        return 'B+'
    elif percentage>=80:
        return 'B'
    elif percentage>=75:
        return 'C'
    else:
        return 'Fail'

df['Grade']=df['Percentage'].apply(grades)
```
